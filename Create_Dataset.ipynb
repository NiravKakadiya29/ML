```python
df = pd.DataFrame()
```

```python
for i in range(1,501):
    response= requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=aa7c7df970374dbfe509cb48f936b0d5&language=en-US&page={i}')
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','original_language','vote_average','vote_count','adult']]
    df = pd.concat([df, temp_df], ignore_index=True)
    
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?id=1Qe463-azf819qhpiZgX9vxisWxSQeCys'
df = pd.read_csv(url,header=None,usecols=[0,1,2])
df.columns = ['Class lable','Alcohol','Malic acid']
```

```python
df.head()
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict={1:'red',2:'blue',3:'green'}
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['Class lable'],palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop('Class lable',axis=1),df['Class lable'],test_size=0.3,random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
fig, (ax1,ax2)= plt.subplots(ncols=2,figsize=(12,6))

# before
ax1.scatter(X_train['Alcohol'],X_train['Malic acid'],c=y_train)
ax1.set_xlabel('Alcohol')
ax1.set_ylabel('Malic acid')
ax1.set_title('Before Scaling')

# after
ax2.scatter(X_train_scaled['Alcohol'],X_train_scaled['Malic acid'],c=y_train)
ax2.set_xlabel('Alcohol')
ax2.set_ylabel('Malic acid')
ax2.set_title('After Scaling')
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,6))

# before

ax1.set_xlabel('Alcohol')
ax1.set_ylabel('Density')
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alcohol'],ax=ax1)
sns.kdeplot(X_train['Malic acid'],ax=ax1)

# after

ax2.set_xlabel('Alcohol')
ax2.set_ylabel('Density')
ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(X_train_scaled['Malic acid'],ax=ax2)
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,6))

# before
ax1.set_xlabel('Alcohol')
ax1.set_ylabel('Density')
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alcohol'],ax=ax1)

# after
ax2.set_xlabel('Alcohol')
ax2.set_ylabel('Density')
ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['Alcohol'],ax=ax2)
plt.show()
```

```python
new_data = car.drop(['Car_Name'],axis=1)
new_data['Current_year']=2025
new_data.head()
```

```python
new_data['Number_of_year']= new_data['Current_year'] - new_data['Year']
new_data.head()
```

```python
new_data.drop(['Year'],axis = 1,inplace = True)
new_data.head()
```

```python
new_data = pd.get_dummies(new_data,drop_first=True)
```

```python
new_data.head()
```

```python
new_data = new_data.drop(['Current_year'],axis=1)
new_data.head()
```

```python
X = new_data.iloc[:,1:]
y = new_data.iloc[:, 0]
```

```python
X.head()
```

```python
print(X.shape)  # (rows, number of features)
print(y.shape)  # (rows,)
```

```python
from sklearn.ensemble import ExtraTreesRegressor

model = ExtraTreesRegressor()
model.fit(X, y)
```

```python
print(model.feature_importances_)
```

```python
feature_importances = pd.Series(model.feature_importances_, index=X.columns)
feature_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = RandomForestRegressor()
```

```python
# # Define the parameter grid
# param_dist = {
#     'n_estimators': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200],
#     'max_depth': [5, 10, 15, 20, 25, 30],
#     'max_features': ['sqrt'],
#     'min_samples_leaf': [1, 2, 5, 10],  # Corrected name
#     'min_samples_split': [2, 5, 10, 15, 100]  # Corrected name
# }

random_grid = {'n_estimators': [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)],
               'max_features': ['sqrt','log2',None],
               'max_depth': [int(x) for x in np.linspace(5, 30, num = 6)],
               'min_samples_split': [2, 5, 10, 15, 100],
               'min_samples_leaf': [1, 2, 5, 10]
            }
print(random_grid)
```

```python
# # Define RandomizedSearchCV
# random_search = RandomizedSearchCV(
#     estimator=rf,  # Directly pass the classifier
#     param_distributions=param_dist,
#     n_iter=10,  # Adjust as needed
#     cv=5,
#     verbose=2,
#     random_state=42,
#     scoring='neg_mean_squared_error',
#     n_jobs=-1  # Use all available processors
# )

model=RandomizedSearchCV(
    estimator = model, 
    param_distributions = random_grid,
    scoring='neg_mean_squared_error', 
    n_iter = 10, cv = 5, verbose=2, 
    random_state=42, n_jobs = -1
)
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
data_frame = {
    'Actual_Value': y_test,
    'Prediction_Value':prediction
}

pred_vs_actu = pd.DataFrame(data_frame,columns=data_frame.keys())
pred_vs_actu.head()
```

```python
plt.figure(figsize=(5,3))
```

```python
sns.displot(y_test-prediction)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, prediction)
mse = mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, prediction)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R):", r2)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(n_estimators=500, learning_rate=0.05, random_state=42)
xgb.fit(X_train, y_train)

xgb_pred = xgb.predict(X_test)

print("XGBoost R2 Score:", r2_score(y_test, xgb_pred))
```
